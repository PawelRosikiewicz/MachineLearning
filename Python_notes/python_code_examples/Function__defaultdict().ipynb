```python
"""

    ________________________

        defaultdict()
    ________________________


    1.     What it is 
           defaultdict = list with tupples [("key", <value>)]
                         Value = can be = list, set, int,  depending on default factory

    2.    How to use defaultdict()?

          First, you must initialize defaultdict() and give one of several 
          options calles default_factory that will chnage the bachaviopur of that object (here called as d)

          default_factory   
              <name>         <code example>
               list   ; d = defaultdict(list)  ; used to collect values associated with unique items in a list (see eg.1)
               int    ; d = defaultdict(int)   ; to countn unique items (see eg. 2)
               set.   ; d = defaultdict(set)   ; to collect values associated with unique items in a set, ie they also will be unique !!! (see eg.3)

"""

# EXAMPLES,


### LIST -----------------------------------------------------------------------
###      Used to find unique values in list and values associated with them


# eg. 1
s = [('yellow', 1), ('blue', 2), ('yellow', 3), ('blue', 4), ('red', 1)]
d = defaultdict(list)
for k, v in s:
    d[k].append(v)
    
d.items()
#[('blue', [2, 4]), ('red', [1]), ('yellow', [1, 3])] 

#.  type(d) # collections.defaultdict


### INT -----------------------------------------------------------------------
###     Used to count unique values in a list/string


# eg.2
s = 'mississippi'
d = defaultdict(int)
for k in s:
     d[k] += 1

d.items()
# [('i', 4), ('p', 2), ('s', 4), ('m', 1)]


### SET -----------------------------------------------------------------------
###     Used to collect values associated with unique items in a set, ie they also will be uniqu


# eg. 3
s = [('red', 1), ('blue', 2), ('red', 3), ('blue', 4), ('red', 1), ('blue', 4)]
d = defaultdict(set)
for k, v in s:
    print(k, v)
    d[k].add(v)
        
d.items()
# [('blue', set([2, 4])), ('red', set([1, 3]))]
```
