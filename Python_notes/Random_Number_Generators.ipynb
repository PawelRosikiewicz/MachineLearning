```python
"""
    Trick 1
    
    create series of normally distributed data 
    with different means eg: 1,2,3,4...n
"""

import numpy as np
import pandas as pd
import scipy.stats as stats 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


#### use np.add

# ... two integers will be added
np.add(1, 4)
      # 5

# ... array + vector, 
#.    a vector is 
#.    broadcasted and then each element added 
#.    to all element in a correponding column will be 
x1 = np.zeros([3,3])
x2 = np.arange(3)
np.add(x1, x2)
      # array([[0., 1., 2.],
      #        [0., 1., 2.],
      #        [0., 1., 2.]])

    
#### plot    
    
# data for plot
x1 = stats.norm.rvs(size=[1000, 3]) # arr, (1000, 3)
x2 = np.arange(9,step=3)
x3 = np.add(x1, x2)

# plot
sns.set_context("poster")
fig = plt.figure(figsize=(4,4))
for i in range(3):
    plt.hist(x3[:,i], alpha=0.8)
sns.despine()
plt.title(f"""array with samples take from norm. distr
              with different means""", fontsize=15)
```
