```python
"""
    PYTHON SETS - DATA STRUCTURE THAT WAS NOT DESCRIBED BEFORE
"""

# ---------------------------------------------------------
# New sets

# create sets
set1 = {"a", "b", "c"} # takes only unique values
set2 = set(["c", "b", "f"]) # from list
set3 = {*set1, *set2} # unpacking serts with "*"
set4 = set1.copy() # copy

# LINK: https://docs.python.org/2/library/sets.html


# ---------------------------------------------------------
# frozenset; like a set but immutable
fset1 = frozenset({"a", "b", "c"})
fset2 = frozenset(set2)
hash(fset1) # only frozensets can be hashed !

# ---------------------------------------------------------
# Inspect and compare sets

# inspect and modify
len(set1)

# compare sets with in/not in; issub... and issuperset
set1 <= set2 ; set1.issubset(set2) # False # test whether every element in sset1 is in set2
set1 >= set2 ; set1.issuperset(set2) # False # test whether every element in set2 is in set1
set1 in set2 # False # test set1 for membership in set2
set1 not in set2 # True # as above but not in,

# return union, intersection, and difference
set1 | set2 ; set1.union(set2) # Union
set1 & set2 ; set2.intersection(set1, set3) # {'b', 'c'} # intersection; can one or more sets in brackets
set1 - set2 ; set1.difference(set2) # {'a'} # set1 - set2
set1 ^ set2 ; set1.symmetric_difference(set2) # {'a', 'f'} # returns set with unique elements of set1 and 2, all common are removed


# ---------------------------------------------------------
# Operations that moddify original set

# modify set elements
set1.add("z")
set1.remove("z") # raises KeyError if not present
set1.discard("z") # removes an item,if present, no keyerror!
set3.clear()
set1.pop() # remove and return last or an arbitrary element from set1; KeyError if set1 is empty

# compare sets and _update its elements == operations used only for sets not, immutableSets
set1 |= set2 ; set1.update(set2) # {'a', 'b', 'c', 'f'} # set1 with elements added from set2; DO NOT USE, better use update()
set1 &= set2 ; set1.intersection_update(set2) # {'b', 'c'} # returns set1 wiht elements also found only in set2
set1 -= set2 ; set1.difference_update(set2) # {"a"} # 
set1 ^= set2 ; #set1.symmetric_difference_update(set2) # {'a', 'f'} # 

# ---------------------------------------------------------
# Iterastion on ultiple sets wiht zip()
for i, oneSet in zip(["set1", "set2", "set3", "set4"], [set1, set2, set3, set4]):
    if "a" in oneSet:
        print(f"a was found in {i} ", oneSet)
    else:
        print(f"a was not found in {i}", oneSet)
        
# zip() vs izip()
# - zip() creates a list of tooples, with items from all list placed in a zipped list
# - izip() creates and iterator, thus no new objects in a memory
#   thus, use itertools.izip(), or itertools.izip_longest(), when managing large objects!

# Two types of izip
# - izip() - finishes on the shortes list
# - izip_longest() - finishes on the longest list, and yields a tuple with None value for lists that already exchaused
```

```python
"""
    Set - examples
"""

# ------------------------------------
# eg1. HR managment by sets

# original sets
engineers = set(['John', 'Jane', 'Jack', 'Janice'])
programmers = set(['Jack', 'Sam', 'Susan', 'Janice'])
managers = set(['Jane', 'Jack', 'Susan', 'Zack'])
engineers.add('Marvin') 

# new sets
employees = engineers | programmers | managers           # union
engineering_management = engineers & managers            # intersection
fulltime_management = managers - engineers - programmers # difference

# testting
employees.issuperset(engineers)  
employees.update(engineers) 

# For loops on list with sets
for group in [engineers, programmers, managers, employees]: # doctest: +SKIP
    group.discard('Susan')          # unconditionally remove element
    print(group)
```
