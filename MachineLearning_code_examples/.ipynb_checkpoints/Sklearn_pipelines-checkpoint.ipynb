### lest come back to data processing piepeline - simple example
* the goals is to use 30 different features to predict whther someone has cancer or not.

```python
# load example dataset
data = load_breast_cancer()
X, y = data.data, data.target

# split to train/test
X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, train_size=0.7, random_state=0) 

# scale input data
scaler = preprocessing.StandardScaler().fit(X_tr)  
X_tr = scaler.transform(X_tr)  
X_te = scaler.transform(X_te)  

# Create classifier & train it
clf = neighbors.KNeighborsClassifier(n_neighbors=2)
clf.fit(X_tr, y_tr)  

# predict test values and check summary
y_pred = clf.predict(X_te)  
print(f'acc = {np.round(accuracy_score(y_te, y_pred), 2)}')

# create confusion matrix
plot_confusion_matrix(X_te, y_te, clf, fullinfo=True)
```

```python
from sklearn.metrics import confusion_matrix
from itertools import combinations

# Function, ...................................................
def plot_confusion_matrix(X, y, model, fullinfo=False):
    '''creates annotated heatmap for confusion matrix
        . X; 1D or 2D array, with input data
        . y; numpy vector with target variable
        . model; trained model, or pipeline
        . full_info; displays nr and % of each class, 
    '''
    matrix = confusion_matrix(y_true=y_te, y_pred=model.predict(X_te))
    matrix = pd.DataFrame(
        matrix,
        columns=[f'pred:{x}' for x in range(matrix.shape[1])],
        index=[f'True:{x}' for x in range(matrix.shape[0])]
    )
    # calulate percentate of each True class
    matrix_perc = matrix/matrix.sum()*100
    print(matrix_perc)
    
    # plot the results with annotated heatmap 
    plt.figure(figsize = (4,3))
    if fullinfo==True: 
        # create array, with labels, same dims as X
        labels = []
        for comb in combinations(range(matrix.shape[0]), 2):
            print(matrix.iloc[comb[0], comb[1]])
            labels.append(
                [f'{matrix.iloc[comb[0], comb[1]]}\n{np.round(matrix_perc.iloc[comb[0], comb[1]], 0)}%']
            )
            print(labels)
        labels = np.array(labels).reshape(matrix.shape[0],matrix.shape[0])        
        sns.heatmap(matrix, annot=labels, fmt='', cmap='Blues')    
        
    elif fullinfo==False:
        sns.heatmap(matrix/matrix.sum(), fmt=".2%",
                annot=True, cmap="Reds")

    plt.show();
    
# create confusion matrix
plot_confusion_matrix(X_te, y_te, clf, fullinfo=True)
```

despite having relatively high accuracy ~90%, our model, miscalssify almost 16% of sick patients with cancer, and no cancer (pred:0). we may try to improve our model, using different echniques.  but first we may want to place preprocessing steps and the model in skleanr pipeline that can be run using standard sklearn API methods fit, predict, score etc...


### place elements of the piepline in sklearn pipeline fucntion

(1) encapsulate multiple steps with pipeline
https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html

```python
from sklearn.pipeline import Pipeline

# each step requires a name, and a function
pipe = Pipeline([
 	   ('scaler', preprocessing.StandardScaler()),
	   ('knn',neighbors.KNeighborsClassifier(n_neighbors=5))
	])
```

```python
# visualize the pipeline
from sklearn import set_config
set_config(display="diagram"); pipe
# beatiful visualizations (HTML)
```

```python
# to see parameters at each step
pipe.get_params() 

# basic text info
pipe.named_steps
```

(2) reuse that with different k-values

```python
# test different k values
results = []
for k in list(range(2,30)):
    # set k in the pipeline
    pipe.set_params(knn__n_neighbors=k)
    pipe.fit(X_tr, y_tr)
    
    # collect the results
    results.append({
        'k':k,
        'train_acc': accuracy_score(y_tr, pipe.predict(X_tr)),
        'test_acc': accuracy_score(y_te, pipe.predict(X_te))
    })
    
# convert the results to pd.dataframe & plot
gs = pd.DataFrame(results)

# plot results
plt.style.use("ggplot")
plt.plot(gs.loc[:,'k'], gs.loc[:,'train_acc'], label="train acc")
plt.plot(gs.loc[:,'k'], gs.loc[:,'test_acc'], label="test acc")
plt.xlabel("k")
plt.ylabel("accuracy")
plt.legend()   
plt.show()

```
