```python
import numpy as np
import pandas as pd

# data and models
from sklearn.datasets import make_blobs
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# curves
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc

# stats
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

# for plots
import matplotlib.pyplot as plt
import seaborn as sns
```

## Example 1. ROC curve for logistic regression
---

#### create dataset wqith 2 classes, 20 features and 1000 samples 

```python
# generate 2-class dataset
""" by default, its 20 features, 10 are informative,  
    make_regression() - is a similar funciton for regression models
    make_blobs() - Generate isotropic Gaussian blobs for clustering
"""

# use make_regression
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
print(X.shape)

# use make_blobs
X, y = make_blobs(
    n_samples=10000,  n_features=20, centers=2, random_state=2, cluster_std=10)

for i in range(2):
    plt.scatter(X[y==i,0], X[y==i,1], s=1)
```

#### split the data, fit logreg model, and predict proba. for test data

```python
# split into train/test sets
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.5, random_state=2)

# fit a model
model = LogisticRegression(solver='lbfgs')
model.fit(trainX, trainy)

# predict probabilities & 
"keep probabilities for the positive outcome only"
lr_probs = model.predict_proba(testX)
lr_probs = lr_probs[:, 1]

# generate a no skill prediction (majority class), 
"any value will do as long as it is the same for all targets"
ns_probs = [1 for _ in range(len(testy))]
```

#### Plot ROC curve

```python
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(testy, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(testy, lr_probs)

# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')

# labels & legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

#### AUC

```python
# calculate scores
ns_auc = roc_auc_score(testy, ns_probs)
lr_auc = roc_auc_score(testy, lr_probs)

# print
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Logistic: ROC AUC=%.3f' % (lr_auc))
```

## 2. PRECISION-RECAL CURVE

#### first, use fitted model to predict classes, for precision scores

```python
# predict class values
yhat = model.predict(testX)

# generate data for model predicting only one class
no_skill = len(testy[testy==1]) / len(testy)

# cacluate pr curve
lr_precision, lr_recall, _ = precision_recall_curve(testy, lr_probs)
```

#### plot the precision-recall curves

```python
# plot the precision-recall curves
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
plt.plot(lr_recall, lr_precision, marker='.', label='Logistic')

# labels & legend
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()
```

#### AUC and F1 scores

```python
lr_f1, lr_auc = f1_score(testy, yhat), auc(lr_recall, lr_precision)
print('Logistic: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
```

## Part 2. We will compare different models on balanced and imbalanced datasets using ROC curves and Precision-Recal Curves

### step 1. we will define two functions used for creating the data for each example, and drawing curves with results from different models

```python
# define function that creates, and prepare the data
def create_data(
    bias=0, bias_class=0, test_size=0.5, plot=False, blobs_dct=None, 
    split=True):
    """ 
        generate 2-class dataset with make_blobs()
        by default, its 20 features, 10 are informative
        ------------------
        . bias;             how many samples to remove from one of two classes
                            caution, it must be between 0, and 1 
        . bias_class;       underrepresented class_name to remove samples if bias!=0
        . blobs_dct;        dct with params for make_blobs
        . test_size;        test_size for train_test_split()
        . split;            if True, it will resturn train and test data, 
                            with X,y for each of them
    """

    if blobs_dct is None:
        blobs_dct= dict(
            n_samples=10000, n_features=20, n_classes=2, random_state=1)
    else:
        pass

    # use make_blobs
    X, y = make_classification(**blobs_dct)
    
    # remove samples from one class to createbiased dataset
    class_idx = np.where(y==bias_class)[0]

    if bias!=0 and bias<1:
        class_idx = np.random.choice(
            class_idx, size=int(class_idx.shape[0]*bias), replace=False)
        X = np.delete(X, class_idx, 0)
        y = np.delete(y, class_idx)
    elif bias==1:
        X = np.delete(X, class_idx, 0)
        y = np.delete(y, class_idx)        
    else:
        pass
        
    # count samples in eahc class
    unique, counts = np.unique(y, return_counts=True)

    # plot
    if plot==True:
        for i in range(2):
            plt.scatter(X[y==i,0], X[y==i,1], s=1)
        plt.title(dict(zip(unique, counts)))
        sns.despine()
        plt.show();
    else:
        pass
    
    if split==True:
        # split data into train/test sets & return
        trainX, testX, trainy, testy = train_test_split(
            X, y, test_size=test_size, random_state=2)
        return trainX, testX, trainy, testy
    else:
        return X, y
        
```

```python
### function; train models, and plotting ROC and Precision-Recal Curves
def train_models(
    trainX, testX, trainy, testy, models, 
    plot_title="", figsize=(12,5), positive_class=1):
    """
        This function, train provided models
        subsequently, it plots ROC and Precision/Recal curves for these models
        and return a dictionary with statistics, such as ROc AUC and F1 for each of them
        ------------------
        . trainX, testX, trainy, testy;  imput data provided by create_data()
        . models; dict, key: model_name, value: skleanr model with provided parameters, 
        . plot_title_; str
        . figsize; tuple, eg (12,5) for plot size 
        . positive_class; int, 0, or 1, class name used as positive in confucion matrices
        
    """
    
    results=[]
    
    # create plot with 2 subplots for ROC and PR curves
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=figsize)
    
    i=0
    for model_name, model in models.items(): 
        # fit a model
        model.fit(trainX, trainy)
    
        # predict probabilities & 
        "keep probabilities for the positive outcome only"
        lr_probs = model.predict_proba(testX)
        lr_probs = lr_probs[:, 1]

        # predict class values
        yhat = model.predict(testX)         
    
    
        # .. data fpr ROC curve
        # first generate a no skill prediction (majority class), 
        "any value will do as long as it is the same for all targets"
        ns_probs = [positive_class for _ in range(len(testy))]          
        
        # calculate roc curves
        ns_fpr, ns_tpr, _ = roc_curve(testy, ns_probs)
        lr_fpr, lr_tpr, _ = roc_curve(testy, lr_probs)   
        
    
        # .. data for precision-recall curve
        # cacluate pr curve
        lr_precision, lr_recall, _ = precision_recall_curve(testy, lr_probs)    
        
        # generate data for model predicting only one class (positive class)
        " its == to frequency of the class labelled as positive, typically the most frequent"
        no_skill = len(testy[testy==positive_class])/len(testy) 
        
        # .. plots
        # plot the roc curve for the model
        axs[0].plot(lr_fpr, lr_tpr, marker='.', label=model_name)

        # plot the precision-recall curves
        axs[1].plot(lr_recall, lr_precision, marker='.', label=model_name)

        # add model predicting one class only
        if i==len(models)-1:
            axs[0].plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
            axs[1].plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
        else:
            pass        
        i+=1

        # count samples in eahc class
        unique, counts = np.unique(testy, return_counts=True) 

        # collect stats
        results.append({
            f"model": model_name,
            "ROC AUC": roc_auc_score(testy, lr_probs),
            "PRC AUC":  auc(lr_recall, lr_precision),
            "Recall": recall_score(testy, yhat),
            "Presision": precision_score(testy, yhat), 
            "F1": f1_score(testy, yhat),
            "counts for test data": dict(zip(unique, counts))  
        })
        
    # labels & legend      
    axs[0].set_ylim(0,1)
    axs[0].set_title("ROC curve")
    axs[0].set_xlabel('False Positive Rate  FP/(FP+TN)')
    axs[0].set_ylabel('True Positive Rate TP/(TP+FN)')
    axs[0].legend()      

    axs[1].set_ylim(0,1)
    axs[1].set_title("Precision-Recall curve")    
    axs[1].set_xlabel('Recall TP/(TP+FN)')
    axs[1].set_ylabel('Precision TP/(TP+FP)')
    axs[1].legend()     

    sns.despine()
    fig.tight_layout()
    plt.show();  

    return pd.DataFrame(results)
```

## PART 3. Example showing how ROC curves can be used to calibrate the threshold with traned models

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

### Step 1. Create pipeline to find optimal hyperparameters for knn model

```python
# search optimal k-value with gridsearch and corss-validation
def knn_grid_search(Xdata, ydata, pipe, klist=[2], N=3):
    '''searches optimal k-value with gridsearch and corss-validation'''
    
    # .. grid search
    results=[]
    for run_idx in range(N):
        for i,k in enumerate(klist):
            # split the data
            X_tr, X_val, y_tr, y_val = train_test_split(
                Xdata, ydata, test_size=0.33, random_state=run_idx)

            # set_params in the pipe
            pipe.set_params(knn__n_neighbors=k)
            pipe.fit(X_tr, y_tr)

            # save the results
            results.append({
                "k":k,
                "run_id":run_idx,
                "Acc_tr":pipe.score(X_tr,y_tr),
                "Acc_valid":pipe.score(X_val, y_val)
            })
    
    # groupby
    df_results = pd.DataFrame(results)   
    return df_results
```

```python
def plot_knn_gs_resuls(df_results):
    # .. summarize the results, 
    # find all used k-values
    klist = df_results.k.unique().tolist()
    
    # groupby
    df_results = pd.DataFrame(df_results)
    grp = df_results.groupby("k")
    
    # find means±sd
    Acc_means_tr  = grp.Acc_tr.mean()
    Acc_std_tr    = grp.Acc_tr.std()
    Acc_means_val = grp.Acc_valid.mean()
    Acc_std_val   = grp.Acc_valid.std()
    
    # .. plot, 
    # plot results
    fig, ax = plt.subplots(nrows=1, ncols=1)
    ax.plot(klist, Acc_means_tr, label="train Acc")
    ax.plot(klist, Acc_means_val, label="validation Acc")
    
    # find and plot "best performing k value"
    best_performing_k_idx = Acc_means_val.idxmax()
    ax.scatter(best_performing_k_idx, Acc_means_val.max(), 
               marker="*", color="blue", zorder=10)
    
    # add std's
    ax.fill_between(klist, Acc_means_tr-Acc_std_tr, Acc_means_tr+Acc_std_tr, alpha=0.3)
    ax.fill_between(klist, Acc_means_val-Acc_std_val, Acc_means_val+Acc_std_val, alpha=0.3)
    
    # axes and labels, 
    ax.set_title(f"best performing k: {best_performing_k_idx}")
    ax.set_xlabel("k")
    ax.set_ylabel("Accuracy")
    ax.legend()
    sns.despine()
    plt.show()
    
    return best_performing_k_idx
```

### Step 2. we will use ROC and Precision-Recall curves to find the most optimal theshold that allows the highes sensitivety and specificity of our predictions

The default threshold for knn models, with 2 classes is 0.5, 



"""
    Goal: to increase the number of times we detect sick patients, 
          at the cost of misclassyfying heathly patients
          
    Method:
          Firs, we will use ROC (Receiver Operating Characteristic)
          to find how small the threshold should be to increate the sensitiviy,
          without reducing the speficity too much (specificity = TN/(TN+FP))
          We will also use ROC Curve to see whther the improvement is even possible, 
          and how bad will be the FPR (FP/(FP+TN)) when manipulating the thresholds, 
          associated with each value, 
          
    Comments:
          - the default threshold with 2 classes is 0.5, 
          - we can also play with distance funciton, 
          - feature space
          - and decision function, that in that case ywoudl be simply the threshold,

"""



```python
# create the data
X, y = create_data(bias=0, bias_class=0, plot=False, split=False)

# define the pipeline
pipe = Pipeline([
    ("scaler", StandardScaler()), # z = (x - u) / s
    ("knn",KNeighborsClassifier(
        algorithm="brute",
        n_jobs=-1
        ))
    ])

# grid search and plot
df_results = knn_grid_search(Xdata=trainX, ydata=trainy, pipe=pipe, klist=list(range(1,51)))
best_k = plot_knn_gs_resuls(df_results)  
```

```python
# lest resuse fucntiosn that we created in part 2. 
#.     but instead of models, we will provide pipeline
pipe.set_params(knn__n_neighbors=31)

# split data into train/test sets & return
"""here we just split data once, buy you can do similar cross-validation 
    as it was done for hyperarameter k search to find optimal threshold
"""
trainX, testX, trainy, testy = train_test_split(
    X, y, test_size=0.3, random_state=1)

results = train_models(trainX=trainX, testX=testX, 
             trainy=trainy, testy=testy, 
             models=dict(knn_pipeline=pipe)
)

pd.DataFrame(results)
```

see what are the avaiable thresholds

```python
# Compute probabilities
y_te_probs = pipe.predict_proba(testX)

# fond unique values of probabilities assigned to class 1
unique, counts = np.unique(y_te_probs[1,], return_counts=True)
unique
```

### finally, we can check, recal and fpr for each threshold, and we can decide if the threshold shoudl be modified. 

```python
# ROC curve
fpr, tpr, thresholds = roc_curve(
    y_true    = testy, 
    y_score   = y_te_probs[:, 1]
)

# place the results in df, 
roc_results = pd.DataFrame({
    'thresholds': thresholds,
    'fpr': fpr,                # x-axis: false positive rate: FDR = FP/(FP+TN)
    'tpr recal': tpr,       # y-axis: Recal, TP rate; TPR = TP/(TP+FN)
})

roc_results.sort_values(by="thresholds")
```

### we decided to lower the threshold, to 0.3,
- because it will allow detecting 95% of positive cases,     
- However, it will increase the number of false posives 4x, from 6% to 24-25%    

HOW TO DO THAT?
- the easiest way, is to write small function, that uses fitted model or pipeline, to predict probabilities, and then to use new threshold to provide your custom predicitons
- here is a post on that topic with more details https://stackoverflow.com/questions/19984957/scikit-learn-predict-default-threshold

lest to that:)

```python
def predict_class(X, trained_model, tr, class_names=(0,1)):
    '''predicts binary class wiht trained model, and custom threshold
        . X; np. array, 
        . trainned_model; skleanr model, or pipeline, fitted on the data
        . tr; cuoff, float, 0-1
        . class_names, tupple, with class names, [0] for class <tr, 
    '''
    probs = trained_model.predict_proba(X)
    return [class_names[1] if p else class_names[0] for p in (probs[:, 1] >= tr)]
```

#### now test our new function andf generate classification report
* as we predicted, we increased recall for class 1, 
* unformtunatelly, at the same time, recall for class 0, and precision for noth classes were decreased, because the number of FP, increased,
* it is especially problemati with skew class composition

```python
custom_predictions = predict_class(testX, pipe, tr=0.3)
classif_report_2 = classification_report(
    y_true=testy, 
    y_pred=custom_predictions
) 

print(classif_report_2)
```
